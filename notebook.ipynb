```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv("data.csv")
data["fraction"] = data["unblocked_sites"] / data["total_sites"]
data = data.drop(columns=["edge_size","unblocked_sites","total_sites"])
```

```python
sns.histplot(data=data["fraction"],stat="count",element="bars")
plt.show()
```

```python
sns.histplot(data=data["fraction"],stat="probability",binrange=[0,1],cumulative=True)
plt.show()
```

```python
x = data["fraction"]
Averages = {"minimum":np.amin(x),"maximum":np.amax(x),"median":np.median(x),"mean":np.mean(data["fraction"]),"standard deviation":np.std(x),"variance":np.var(x)}
Averages
```
